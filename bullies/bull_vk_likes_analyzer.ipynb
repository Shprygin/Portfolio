В данном ноутбуке я определю топ лайкеров постов юзеров до 2018 (вкл.) года. В дальнейшем этих пользователей будем проверять на критерии буллинга в другом ноутбуке

```python
import vk
import pandas as pd
session = vk.Session(access_token='5fb8befe5fb8befe5fb8befe465fc2e60955fb85fb8befe3e135e3d836bfd885469434b') 
api = vk.API(session)
```

```python
member_ids = api.groups.getMembers(group_id=147814667, v=5.131)['items']  #список ша пользователей
```

```python
member_pages = []
for id_ in member_ids:
    member_pages.append('https://vk.com/id'+str(id_))
member_pages #список ссылок на пользователей вк
```

```python
from datetime import datetime
```

```python
walls = [] #список стен пользователей, на каждой стене список постов
open_ids = []
locked_ids=[]

for id_ in member_ids:
    try:
        walls.append(api.wall.get(owner_id=id_, count = 550, v=5.131)['items'])
        open_ids.append(id_)
    except:
        locked_ids.append(id_)
        pass
```

```python
activity_network = dict.fromkeys(open_ids)
activity_network
```

```python
order = 0
for wall in walls:
    post_links = []
    for post in wall:
        if  datetime.utcfromtimestamp(int(post['date'])).year <= 2018: #тк группа была активна до 18го года, то значит и абьюз был до этого года включительно
            post_links.append(post['id'])
    activity_network[open_ids[order]] = post_links
    order += 1  
```

```python
for i in list(activity_network.keys()): #удалил айди без постов
    if activity_network[i] == []:
        del activity_network[i]

activity_network
```

```python
activity_network_likers = dict.fromkeys(activity_network.keys())
activity_network_likers
```

```python
activity_network_likers
for owner_id in list(activity_network.keys()): #удалил айди без постов
    likers = []
    for item_id in activity_network[owner_id]:
        likers += api.likes.getList(type='post', item_id=item_id, owner_id=owner_id, v=5.131)['items']
        sleep(1)
    activity_network_likers[owner_id] = likers
activity_network_likers[11346734]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
for user in list(activity_network_likers.keys()):
    string=[]
    for value in activity_network_likers[user]:
        string.append(str(value))
    activity_network_likers[user] = string
activity_network_likers
```

```python
from collections import Counter
```

```python
activity_network_likers_top = activity_network_likers.copy()
for user in list(activity_network_likers_top.keys()):
    c = Counter(activity_network_likers[user]).most_common(3)
    activity_network_likers_top[user]=c
activity_network_likers_top
```
